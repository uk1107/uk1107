```python
#11-A
class Unionfind:
    def __init__(self,n):
        self.parent = [i for i in range(n)]
        self .height = [0 for i in range(n)]

    def get_root(self, i):
        if self.parent[i] == i:
            return i
        else:
            self.parent[i] = self.get_root(self.parent[i])
            return self.parent[i]

    def unite(self,i,j):
        root_i = self.get_root(i)
        root_j = self.get_root(j)
        if root_i != root_j:
            if self.height[root_i] <  self. height[root_j]:
                self.parent[root_i] = root_j
            else:
                self.parent[root_j] = root_i
                if self.height[root_i] == self.height[root_j]:
                    self.height[root_i] += 1
    def is_in_group(self,i ,j):
        if self.get_root(i) == self.get_root(j):
            return True
        else:
            return False

N,M = map(int,input().split())
e_list = []
for i in range(M):
    arr = list(map(int,input().split()))
    e_list.append(arr)

def kruskal(V,e_list):
    e_cost_sorted = []
    for i in e_list:
        e_cost_sorted.append([i[2],i[0],i[1]])
    e_cost_sorted.sort()
    uf_tree = Unionfind(V)
    mst = []
    for i in e_cost_sorted:
        if not uf_tree.is_in_group(i[1],i[2]):
            uf_tree.unite(i[1],i[2])
            mst.append(i[0])
    return sum(mst)

print(kruskal(N,e_list))
```
